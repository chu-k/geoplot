```python
ngrid = 25
viridis = mpl.colormaps['twilight'].resampled(ngrid)
fig, ax = plt.subplots(figsize=(ngrid,ngrid))
# ax.set_xlim((-ngrid,ngrid))
# ax.set_ylim((-ngrid,ngrid))

maxsize = 2

xd = np.linspace(0, ngrid, ngrid)
yd = np.linspace(0, ngrid, ngrid)

for i, x in enumerate(xd):
    deg_rotx = 180.0 * i/ngrid
    for j, y in enumerate(yd):
        tri = gp.triangle.Triangle(np.array((x,y)), maxsize*i/ngrid*j/ngrid)
        deg_roty = -30.0 * j/ngrid
        tri.rotate2d(np.radians(deg_rotx))
        tri.rotate2d(np.radians(deg_roty ))
        x2 = tri.points[:,0]
        y2 = tri.points[:,1]
        ax.fill(list(x2), list(y2), color=viridis(tri.center[0]/ngrid), alpha=0.8)

        # ax.scatter(*tri2.center, c='g')
```
